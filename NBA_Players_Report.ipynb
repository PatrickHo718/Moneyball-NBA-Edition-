# Player Selection Report for the Wizards

From this project, I conducted on the NBA Player Dataset from 2023-2024, I have selected the following players for the Wizards to consider: Tyus Jones, Jalen Williams, and Josh Giddey. My conclusion is based on the players' performance metrics, including total field points, assists, and rebounds that generally applies to all the 5 main positions on a court: Shooting Guard, Point Guard, Power Forward, Small Forward, and Center. In addition, I have also considered their salary to ensure that they fit within the Wizards' budget constraints. 

The method I used for this data science lab was K-Means clustering, which is a common unsupervised learning method used to group data points that are similar to each other. Unsupervised learning is a type of machine learning tool that trains a model without labeled data. In this case, I used K-Means clustering to group players according to the aforementioned metrics, in which the players will be labeled based on which cluster they belong to. 

To determine the optimal number of clusters, I used the elbow method to find the best number of clusters. It works by plotting the sum of squared distance, which is measure that calculates each point's distance to its assigned cluster centroid. The elbow point on the plot is the point where the rate of decrease changes the most, which usually appears to look flatter after that point. From the graph below, you can see that the elbow point is not very obvious as the most of the points still appear to be decreasing at a fast rate. However, I have chosen 4 clusters as the optimal number as there is a good balance between the number of clusters and a lower chance of overfitting the model, where a model learns the training data too well that it cannot generalize.

```python
Image("elbow_method.png", width=400, height=400)
```

To support this decision with the silouette score, I plotted the silhouette score for each cluster. The silhouette score is a measure of how well-separated and similar an object is to its own cluster in comparison with other clusters. The silhouette score that was achieved was 0.43 for 4 clusters, which is a fair score indicating that the clusters are reasonably well-separated.

```python
Image("silhouette_scores.png", width=400, height=400)
```

When the players were grouped into 4 clusters on the graph, there was an interesting pattern that emerged. There was a cluster that contains players that have lower performance metrics on one edge, a cluster in the middle that contains average players, and two clusters on the other edge that have a higher performance metrics. The difference between the two clusters on the higher performance metrics is that one cluster contains players who have better stats in total rebounds, whereas the other cluster contains players who have better stats in assists.

The way I value the players is by looking at their individual performance and their ability to work well with others. That is why I put more weight on assists than rebounds as assists are a better indicator of a player's ability to work with others. I also value players who can score points, so I put more weight on total field points than assists. 

Last but not least, I also considered the players' salary to ensure that they fit within the Wizards' budget constraints. The players I selected are all in the range of 2 million to 5 million dollars, which is a reasonable range for the Wizards to consider. The way I approached is by summing up the standardized values of total field points, assists, and rebounds, and then dividing it by the salary. This way, I can get a better idea of how much value each player is providing for their salary. 

I concluded that players Tyus Jones, Jalen Williams, and Josh Giddey are the best for the Wizards to consider.
